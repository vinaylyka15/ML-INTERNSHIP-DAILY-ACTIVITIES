```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({
    "color":["red","green","blue","green","red"],
    "size":["small","medium","large","medium","small"],
    "city":["hyderabad","delhi","mumbai","delhi","chennai"],
    "target":[23,56,89,90,98]
})
df
```

# label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df["color_label"]=le.fit_transform(df["color"])
df[["color","color_label"]]
```

#  ONE HOT ENCODING

```python
# craetes a new binary column for each category with a 1 indicating the presence of that category and 0 otherwise
#  idesl for nominal data(no inherent order) with low  to medium cardinality, and for linear models or neural networks.
```

```python
one_hot=pd.get_dummies(df["color"],prefix="color")
one_hot
```

```python
oh=pd.get_dummies(df,columns=["color"])
print(oh)
```

#  ordinal encoder

```python
#   similar to label encoding , but the integer assignment is based on the specific,predefined order or rank of the categories
# eg low :0 medium:1,high: 2
# use specifically when the categorical variable has a natural meaningful oeder.
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
order=[["small","medium","large"]]
oe=OrdinalEncoder(categories=order)
df["size_ordinal"]=oe.fit_transform(df[["size"]])
df[["size","size_ordinal"]]
```

# target encoding

```python
# each category is replaced with the mean of the target variable for the category
# effective for hgh cardinality features and when there a strong relationshipbetween and category and the target variable but 
```

```python
te=df.groupby("city")["target"].mean()
df["city_target_enc"]=df["city"].map(te)
df[["city","city_target_enc"]]
```

# frequency encoding

```python
# replace each category with its frequency (count) in thye dataset.
# useful when the frquency of a category is predictive of the target variable.
```

```python
freq=df["city"].count()
df["city-freq_enc"]=df["city"].map(freq)
df[["city","city_freq_enc"]]
```
