# feature selection

```python
# feature selection: the process of identifying  and keeping only the most relevant features while removing irrelevant or redundant ones to improve model performance and reduce overfitting.
# why it remoes noise
# too many features confuse the model , irrelevant features cause overfitting, ,noise reduce prediction accuracy
# types of feature selection
# 1)filter methods(correlation,chi-square),2) wrapper methods(forward,backward,RFE)  3) embedded methods(lasso,tree-based models)
#  example: removing weakly correlated features, dropping features with zero importance , selecting top feature using RFE( recursive feature elimination)
# impact on model : reducces overfitting, improves generalization, speeds up trainning


```
